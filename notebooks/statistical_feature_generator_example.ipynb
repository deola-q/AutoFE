# StatisticalFeatureGenerator Example

Example of unary and pairwise feature generation with selected output.

```python
import sys
from pathlib import Path
import pandas as pd

sys.path.append(str(Path("..") / "AutoFE"))
from stats import StatisticalFeatureGenerator
```

```python
X = pd.DataFrame(
    {
        "income": [1200, 1500, 1100, 1800, 2100, 1700],
        "spend": [500, 700, 400, 900, 1200, 800],
        "age": [23, 30, 22, 45, 37, 29],
    }
)
y = pd.Series([0, 1, 0, 1, 1, 0], name="target")
X
```

```python
gen = StatisticalFeatureGenerator(
    numeric_cols=["income", "spend", "age"],
    unary=["log", "sqrt"],
    pairwise=["ratio", "diff"],
    max_features=6,
)

gen.fit(X, y)
X_out = gen.transform(X)
X_out.head()
```

```python
print("Selected new features:", gen.selected_)
print("Original columns kept:", all(c in X_out.columns for c in X.columns))
print("Output shape:", X_out.shape)
```
