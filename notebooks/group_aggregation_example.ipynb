# GroupAggregationFeatures Example

Minimal runnable example for group-based feature generation.

```python
import sys
from pathlib import Path
import pandas as pd

sys.path.append(str(Path("..") / "AutoFE"))
from stats import GroupAggregationFeatures
```

```python
df = pd.DataFrame(
    {
        "city": ["A", "A", "A", "B", "B", "C"],
        "segment": ["retail", "retail", "corp", "retail", "corp", "corp"],
        "sales": [100, 130, 90, 200, 180, 160],
        "margin": [15, 20, 10, 35, 30, 28],
    }
)
df
```

```python
gen = GroupAggregationFeatures(
    numeric_cols=["sales", "margin"],
    group_cols=["city", "segment"],
    aggs=["mean", "max", "count"],
    add_deviation=True,
    add_rank=True,
)

gen.fit(df)
df_out = gen.transform(df)
df_out.head()
```

```python
print("Generated features:", len(gen.generated_features_))
print(gen.generated_features_[:10])
```
